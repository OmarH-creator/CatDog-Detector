# PetImages Preprocessing Preview

This notebook loads sample images from the `PetImages` folder, converts them to grayscale, resizes them to `128x128`, and displays a few examples from each category.


## 1. Import Libraries

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# Use project-local PetImages folder
DATADIR = os.path.join(os.getcwd(), "PetImages")

CATEGORIES = ["Dog", "Cat"]
IMG_SIZE = 128  # Image size after resizing
```

## 3. Preprocessing Function

```python
def preprocess():
    for category in CATEGORIES:
        path = os.path.join(DATADIR, category)

        count = 0

        for img_name in os.listdir(path):
            img_path = os.path.join(path, img_name)

            # Read image as grayscale to reduce complexity
            img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

            # Resize image to IMG_SIZE x IMG_SIZE
            resized_img = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))

            count += 1

            # Display image
            plt.imshow(resized_img, cmap="gray")
            plt.title(f"{category} sample {count} resized to {IMG_SIZE}x{IMG_SIZE}")
            plt.axis("off")
            plt.show()

            print("Dimensions of the ABOVE image =", resized_img.shape)

            # Show only first 3 valid images per category
            if count == 3:
                break
```

## 4. Run Preprocessing

```python
preprocess()
```
